<font size="6">3.1.2. Hypothesis testing: comparing two groups</font>

<font size="4">1-sample t-test: testing the value of a population mean</font>

```python
# Null hypothesis that mean of VIQ data of all population is 0, 
# The p value is smaller than 0.05 so we reject null hypothesis that mean value of VIQ of population is not 0. 
stats.ttest_1samp(data['VIQ'], 0)
```

<font size="4">2-sample t-test: testing for difference across populations</font>

```python
# We want to check if there is a statistically significant difference between the group of female'viq and male's viq.
# ttest_ind is used calculate the ttest for the means of two independent samples of scores.
# Null hypothesis:  no statistical significance between the viq of female and male.
# Conclusion: P value bigger than 0.05, we can not reject the null hypothesis
female_viq = data[data['Gender'] == 'Female']['VIQ']
male_viq = data[data['Gender'] == 'Male']['VIQ']
stats.ttest_ind(female_viq, male_viq)   
```

<font size="5">3.1.2.2. Paired tests: repeated measurements on the same individuals
</font>

```python
# We want check if there is significantly difference between FSIQ and PIQ. They have identical mean value.
# The p value bigger than 0.05, so we can not reject the null hypothesis.
stats.ttest_ind(data['FSIQ'], data['PIQ'])
```

```python
# ttest_rel is used to calculate the t-test for two related samples. We assume that there exists some 
# links between FSIQ and PIQ, so we use ttest_rel and we found p value is bigger than 0.05, we can not reject the
# the null hypothesis.
stats.ttest_rel(data['FSIQ'], data['PIQ'])   
```

```python
# ttest_1samp is used to calculate the ttest for one group of data.
# Null hypothesis: FSIQ and PIQ have the identical average value.
# The p value is greater than 0.05, we can not reject the null hypothesis.
stats.ttest_1samp(data['FSIQ'] - data['PIQ'], 0)  
```

```python
# Null hypothesis: two related paired samples come from the same distribution.
# Conclusion: p value is greater than 0.05, so we can not reject the null hypothesis.
stats.wilcoxon(data['FSIQ'], data['PIQ'])  
```

<font size="6">3.1.3. Linear models, multiple factors, and analysis of variance</font>

<font size="5">3.1.3.1. formulas to specify statistical models in Python</font>

<font size="4">A simple linear regression</font>

```python
#Return evenly spaced 20 numbers over -5 and 5.
x = np.linspace(-5,5,20)
```

```python
# Reseed the singleton RandomState instance 
np.random.seed(1)
```

```python
# create a linear model with slope 3 and intercept 3+ a random number
y = -5 + 3*x + 4 * np.random.normal(size=x.shape)
# Extract all x and fitted values in linear model
data = pd.DataFrame({'x': x, 'y': y})
data
```

```python
# Use Ordinary Least Squares(OLS) method to linear regression on generated data
from statsmodels.formula.api import ols
model = ols("y ~ x", data).fit()
```

```python
# Detailed information of the ols model. We can find it use the least square and r squred is 0.804 that more than
# 80% of dependent value can be predicted by the independent value.
model.summary()
```

<font size="4">Categorical variables: comparing groups or multiple categories</font>

```python
# Read data from brain size data
# We use OLS model to do linear regression on continuous variable VIQ and categorical variable gender.
data = pd.read_csv("brain_size.csv",sep=';', na_values=".")
model = ols("VIQ ~ Gender + 1", data).fit()
model.summary()
```

```python
# We can use C() to convert a column of variable to be treated as categorical.
model = ols('VIQ ~ C(Gender)', data).fit()
model.summary()
```

```python
# To creat a table to store needed "IQ" values with two types: FSIQ and PIQ 
data_fisq = pd.DataFrame({'iq': data['FSIQ'], 'type': 'fsiq'})
data_piq = pd.DataFrame({'iq': data['PIQ'], 'type': 'piq'})
data_long = pd.concat((data_fisq, data_piq))
data_long
```

```python
# Use ols model to do liner model on two types of iq variable.
model = ols("iq ~ type", data_long).fit()
model.summary()
```

```python
# The p value is greater than 0.05, we can not reject the null hypothesis that there is no significant difference 
# between these two variables
stats.ttest_ind(data['FSIQ'], data['PIQ'])   
```

<font size="5">3.1.3.2. Multiple Regression: including multiple factors</font>

```python
# Read data from iris and generate liner model for multiple factors with z(dependent variabel) = sepal_width 
# and x=name, y = petal_length to find the regression model z = x*c1+y*c2+i+e
data = pd.read_csv('iris.csv')
model = ols('sepal_width ~ name + petal_length', data).fit()
model.summary()
```

<font size="5">3.1.3.3. Post-hoc hypothesis testing: analysis of variance (ANOVA)</font>

```python
# f_test or anova are used to check variance among group means and thus generalizes the t-test beyond two means.
# to check if the two groups of data are significantly different or not
# Null hypothesis: The variance of two samples are equal.
# Conclusion: p value greater than 0.05, we can not reject the null hypothesis.
model.f_test([0, 1, -1, 0])
```

<font size="6">3.1.4. More visualization: seaborn for statistical exploration</font>

```python
# Load data
names = [
    'EDUCATION: Number of years of education',
    'SOUTH: 1=Person lives in South, 0=Person lives elsewhere',
    'SEX: 1=Female, 0=Male',
    'EXPERIENCE: Number of years of work experience',
    'UNION: 1=Union member, 0=Not union member',
    'WAGE: Wage (dollars per hour)',
    'AGE: years',
    'RACE: 1=Other, 2=Hispanic, 3=White',
    'OCCUPATION: 1=Management, 2=Sales, 3=Clerical, 4=Service, 5=Professional, 6=Other',
    'SECTOR: 0=Other, 1=Manufacturing, 2=Construction',
    'MARR: 0=Unmarried,  1=Married',
]
short_names = [n.split(':')[0] for n in names]
data = pd.read_csv('wages.txt', skiprows=27, skipfooter=6, sep=None,
                       header=None)
data.columns = short_names
data
```

```python
data['SEX'] = np.choose(data.SEX, ['male', 'female'])
```

```python
data['WAGE'] = np.log10(data['WAGE'])
```

<font size="6">3.1.5. Testing for interactions</font>

```python
# To check if education would increase more wage on male than female.
# p value of eduation VS gender is 0.056 greater than 0.05, we can not reject the null hypothesis, so the data
# does not support the hypothesis that education would increase wage on female more than male.
result = ols(formula='WAGE ~ EDUCATION + SEX + EDUCATION * SEX',
                data=data).fit()
result.summary()
```
